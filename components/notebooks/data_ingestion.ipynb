```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(
    'Company', axis=1), df.Company, test_size=0.2, random_state=24, stratify=df.Company)
```

```python
X_train.reset_index(inplace=True)
X_train.drop('index', axis=1, inplace=True)
```

```python
X_train['Company'] = y_train.values
X_train.head()
```

```python
X_test.reset_index(inplace=True)
X_test.drop('index', axis=1, inplace=True)
X_test['Company'] = y_test.values
X_test.head()
```

```python
X_train.to_csv(
    "C:/Daya/mlprojects/mlcarproject/artifacts/data/train.csv", index=False)
X_test.to_csv(
    "C:/Daya/mlprojects/mlcarproject/artifacts/data/test.csv", index=False)
```
